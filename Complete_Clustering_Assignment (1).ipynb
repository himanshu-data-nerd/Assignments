# Theoretical Questions
---

## Question 1
What is unsupervised learning in the context of machine learning?

## Answer
Unsupervised learning is a machine learning approach where models are trained on data without labeled outputs. The algorithm identifies hidden structures, patterns, or groupings in the data. Common applications include clustering, dimensionality reduction, and anomaly detection.

---

## Question 2
How does K-Means clustering algorithm work?

## Answer
K-Means clustering partitions data into K clusters by initializing K centroids and assigning each data point to the nearest centroid using a distance metric. Centroids are updated iteratively as the mean of assigned points until convergence.

---

## Question 3
Explain the concept of a dendrogram in hierarchical clustering.

## Answer
A dendrogram is a tree-like diagram that represents the hierarchical merging or splitting of clusters. The vertical axis represents distance or dissimilarity, helping determine the optimal number of clusters.

---

## Question 4
What is the main difference between K-Means and Hierarchical Clustering?

## Answer
K-Means is a centroid-based algorithm requiring a predefined number of clusters, whereas hierarchical clustering creates a nested cluster structure without requiring K beforehand.

---

## Question 5
What are the advantages of DBSCAN over K-Means?

## Answer
DBSCAN can identify arbitrarily shaped clusters, handle noise effectively, and does not require specifying the number of clusters in advance.

---

## Question 6
When would you use Silhouette Score in clustering?

## Answer
Silhouette Score is used after clustering to evaluate cluster quality by measuring cohesion within clusters and separation between clusters.

---

## Question 7
What are the limitations of Hierarchical Clustering?

## Answer
It has high computational complexity, is sensitive to noise, and does not scale well to large datasets.

---

## Question 8
Why is feature scaling important in clustering algorithms like K-Means?

## Answer
K-Means relies on distance calculations. Feature scaling ensures that all features contribute equally to distance computation.

---

## Question 9
How does DBSCAN identify noise points?

## Answer
Points that do not belong to any dense region, defined by minimum samples within epsilon radius, are classified as noise.

---

## Question 10
Define inertia in the context of K-Means.

## Answer
Inertia is the sum of squared distances between each data point and its assigned cluster centroid.

---

## Question 11
What is the elbow method in K-Means clustering?

## Answer
The elbow method identifies the optimal number of clusters by plotting inertia versus K and locating the point where the decrease in inertia slows significantly.

---

## Question 12
Describe the concept of density in DBSCAN.

## Answer
Density refers to the number of data points within a specified neighborhood radius. Clusters are formed in high-density regions.

---

## Question 13
Can hierarchical clustering be used on categorical data?

## Answer
Yes, hierarchical clustering can be applied to categorical data using appropriate distance measures.

---

## Question 14
What does a negative Silhouette Score indicate?

## Answer
A negative score indicates that a data point is closer to another cluster than its assigned cluster.

---

## Question 15
Explain the term linkage criteria in hierarchical clustering.

## Answer
Linkage criteria determine how distances between clusters are calculated during merging.

---

## Question 16
Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

## Answer
K-Means assumes clusters are spherical and equally sized, leading to incorrect clustering when this assumption fails.

---

## Question 17
What are the core parameters in DBSCAN, and how do they influence clustering?

## Answer
The parameters are eps (neighborhood radius) and min_samples (minimum points to form a dense region).

---

## Question 18
How does K-Means++ improve upon standard K-Means initialization?

## Answer
K-Means++ initializes centroids far apart, reducing poor initialization and improving convergence.

---

## Question 19
What is agglomerative clustering?

## Answer
A bottom-up hierarchical clustering method where individual points are merged into clusters iteratively.

---

## Question 20
What makes Silhouette Score a better metric than just inertia for model evaluation?

## Answer
It considers both cluster compactness and separation, unlike inertia which only measures compactness.

## Question 29
Create synthetic data using make_blobs and evaluate silhouette scores for k =2 to 5. Display as a bar chart.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=400, centers=4, random_state=42)

# Store silhouette scores
k_values = range(2, 6)
scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot bar chart
plt.figure(figsize=(6, 4))
plt.bar(k_values, scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k Values")
plt.show()
```

## Question 31
Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters
X, _ = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=2.0,   # large std to create overlap
    random_state=42
)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Create a mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

# Predict cluster for each grid point
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure(figsize=(7, 6))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='X')
plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Question 34
Load the Breast Cancer dataset and compare inertia values for k = 2 to 6 using K-Means. show results in a line plot

```python
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load dataset
data = load_breast_cancer()
X = data.data

# Scale features (important for K-Means)
X_scaled = StandardScaler().fit_transform(X)

# Compute inertia values
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot inertia vs k
plt.figure(figsize=(6, 4))
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method using Breast Cancer Dataset")
plt.show()
```

## Question 44
Load the iris dataset and plot the Sillhouette Coefficient for each sample after KMeans clustering

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
iris = load_iris()
X = iris.data

# Scale features
X_scaled = StandardScaler().fit_transform(X)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Compute silhouette values for each sample
silhouette_vals = silhouette_samples(X_scaled, labels)
avg_score = silhouette_score(X_scaled, labels)

# Plot silhouette coefficients
y_lower = 10
plt.figure(figsize=(7, 5))

for i in range(3):
    cluster_vals = silhouette_vals[labels == i]
    cluster_vals.sort()
    size = cluster_vals.shape[0]
    y_upper = y_lower + size

    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        cluster_vals
    )
    plt.text(-0.05, y_lower + 0.5 * size, f"Cluster {i}")
    y_lower = y_upper + 10

plt.axvline(x=avg_score, linestyle="--")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample Index")
plt.title("Silhouette Plot for Iris Dataset (K-Means)")
plt.show()
```

Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Generate synthetic data with noise
X, _ = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=1.0,
    random_state=42
)

# Add random noise
noise = np.random.uniform(low=-10, high=10, size=(50, 2))
X_noisy = np.vstack([X, noise])

# Scale features
X_scaled = StandardScaler().fit_transform(X_noisy)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count noise points and clusters
noise_count = np.sum(labels == -1)
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print("Number of clusters detected:", num_clusters)
print("Number of noise points detected:", noise_count)
```

## Question 48
Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()
X = digits.data

# Scale features (recommended before t-SNE)
X_scaled = StandardScaler().fit_transform(X)

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_tsne)

# Plot clustering result
plt.figure(figsize=(7, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=30)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
