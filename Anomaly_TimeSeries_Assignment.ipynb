What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

## Answer

# Anomaly Detection

## Introduction
**Anomaly Detection** is a data science and machine learning technique used to identify data points, patterns, or observations that significantly deviate from normal behavior. Such unusual observations are often referred to as *outliers, novelties, or anomalies*. Anomaly detection is widely used in fraud detection, network security, healthcare monitoring, industrial systems, and time-series analysis.


## Definition
An anomaly is an observation that does not conform to the expected pattern of the majority of the data.  
Anomaly detection aims to automatically identify these rare events so that corrective or preventive actions can be taken.


## Importance of Anomaly Detection
- Early detection of system failures  
- Prevention of financial fraud  
- Improved system reliability and safety  
- Reduced operational and maintenance costs  


## Types of Anomalies

Anomalies are broadly classified into three main types:
1. Point Anomalies  
2. Contextual Anomalies  
3. Collective Anomalies  


## 1. Point Anomalies

### Definition
A **point anomaly** occurs when an individual data point is significantly different from the rest of the data, regardless of context.

### Characteristics
- Single observation  
- Easy to detect  
- Most common type of anomaly  

### Examples
- Electricity consumption usually ranges between 200–300 units, but one reading shows 900 units  
- A credit card transaction of ₹5,00,000 when typical spending is below ₹5,000  

### Use Cases
- Fraud detection  
- Sensor fault detection  
- Data quality monitoring  


## 2. Contextual Anomalies

### Definition
A **contextual anomaly** is a data point that is anomalous only within a specific context such as time, location, or environmental condition.

### Characteristics
- Depends on contextual attributes  
- May be normal in a different context  
- Common in time-series data  

### Examples
- Temperature of 30°C is normal in summer but abnormal in winter  
- High electricity usage at 3 AM in a residential area  
- Increased hospital visits during a non-epidemic period  

### Use Cases
- Weather monitoring  
- Energy demand analysis  
- Healthcare surveillance  


## 3. Collective Anomalies

### Definition
A **collective anomaly** occurs when a group of related data points together form an anomalous pattern, even though individual points may appear normal.

### Characteristics
- Sequence or group-based  
- Pattern-level anomaly  
- More difficult to detect  

### Examples
- A series of slightly increased network packets indicating a DDoS attack  
- Gradual decline in machine performance over time  
- Multiple small bank withdrawals indicating money laundering  

### Use Cases
- Network intrusion detection  
- Industrial equipment monitoring  
- Behavioral analysis  


## Comparison of Anomaly Types

| Anomaly Type | Basis of Detection | Example |
|--------------|-------------------|---------|
| Point | Single data point | Unusually high transaction |
| Contextual | Context-dependent | High usage at midnight |
| Collective | Group behavior | Sequence of suspicious events |

## Question 2
Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of approach and use cases.

## Answer
# Comparison of Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)

Anomaly detection algorithms differ in how they define and identify abnormal data points. Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are three widely used techniques with different underlying approaches and use cases.


## 1. Isolation Forest

### Approach
Isolation Forest is a **tree-based, ensemble method** that detects anomalies by randomly partitioning the data.  
Anomalies are isolated faster than normal points because they lie in sparse regions of the feature space and require fewer splits.

### Key Characteristics
- Does not rely on distance or density
- Works well with high-dimensional data
- Efficient and scalable
- Suitable for real-time anomaly detection

### Use Cases
- Fraud detection
- Network intrusion detection
- Real-time monitoring systems
- Large-scale datasets


## 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

### Approach
DBSCAN is a **density-based clustering algorithm** that groups data points into clusters based on neighborhood density.  
Points that do not belong to any dense cluster are labeled as anomalies (noise).

### Key Characteristics
- No need to specify number of clusters
- Can detect arbitrarily shaped clusters
- Sensitive to parameter selection (epsilon, min_samples)
- Less effective in high-dimensional spaces

### Use Cases
- Spatial data analysis
- Noise detection in datasets
- Pattern discovery
- Offline anomaly detection


## 3. Local Outlier Factor (LOF)

### Approach
Local Outlier Factor is a **density-based method** that compares the local density of a data point with the densities of its neighbors.  
Points with significantly lower density than their neighbors are considered anomalies.

### Key Characteristics
- Detects local anomalies
- Sensitive to neighborhood size (k)
- Not ideal for streaming data
- Computationally expensive for large datasets

### Use Cases
- Local anomaly detection
- Regional or segment-based analysis
- Credit card fraud analysis
- Healthcare monitoring

## Comparison Table

| Algorithm | Detection Principle | Strengths | Limitations | Typical Use Cases |
|---------|--------------------|-----------|-------------|-------------------|
| Isolation Forest | Random isolation using trees | Fast, scalable, high-dimensional | Less interpretable | Real-time systems, fraud detection |
| DBSCAN | Density-based clustering | Finds arbitrary shapes, no k needed | Parameter-sensitive, poor in high-dim | Spatial data, noise detection |
| LOF | Local density comparison | Detects local anomalies | Slow, not real-time | Regional anomaly analysis |

## Question 3
What are the key components of a Time Series? Explain each with one example.

## Answer
# Key Components of a Time Series

## Introduction
A **time series** is a sequence of data points recorded at regular time intervals.  
Time series analysis focuses on identifying underlying patterns in the data to understand past behavior and make future predictions.  
A time series is generally composed of four key components: **Trend, Seasonality, Cyclical, and Irregular (Residual

## 1. Trend Component

### Definition
The **trend** represents the long-term movement or direction in the data over an extended period.

### Characteristics
- Shows overall increase or decrease
- Ignores short-term fluctuations
- Reflects long-term growth or decline

### Example
The steady increase in **annual electricity consumption** due to population growth and industrialization.


## 2. Seasonal Component

### Definition
The **seasonal component** captures regular and repeating patterns that occur at fixed intervals within a year.

### Characteristics
- Fixed and predictable
- Repeats over known time periods
- Influenced by calendar effects

### Example
Higher **electricity demand during summer months** due to air-conditioner usage.


## 3. Cyclical Component

### Definition
The **cyclical component** represents long-term oscillations that occur over periods longer than one year and do not have a fixed frequency.

### Characteristics
- Irregular duration
- Related to economic or business cycles
- Hard to predict precisely

### Example
Changes in **power consumption during economic booms and recessions**.


## 4. Irregular (Residual) Component

### Definition
The **irregular component** consists of random, unpredictable variations that cannot be explained by trend, seasonality, or cycles.

### Characteristics
- Random and short-term
- Caused by unexpected events
- Often treated as noise

### Example
A sudden **drop in electricity usage due to a natural disaster or power outage**.


## Summary Table

| Component | Description | Example |
|---------|------------|---------|
| Trend | Long-term direction | Rising annual energy demand |
| Seasonality | Repeating short-term pattern | Summer peak in electricity use |
| Cyclical | Long-term oscillations | Economic expansion and recession |
| Irregular | Random fluctuations | Power outage due to storm |

## Question 4
Define stationarity in time series. How can you test and transform a non-stationary series?

## Answer
# Stationarity in Time Series

## Definition of Stationarity
A time series is said to be **stationary** if its statistical properties remain **constant over time**.  
In a stationary time series:
- The **mean** is constant
- The **variance** is constant
- The **autocovariance** depends only on the lag, not on time

Stationarity is a fundamental assumption in many time series models such as **ARIMA**, as it ensures reliable and stable forecasting.


## Types of Stationarity

### 1. Strict Stationarity
A time series is strictly stationary if the **joint probability distribution** of any set of observations is invariant to time shifts.

> This is a strong condition and rarely tested in practice.


### 2. Weak (Covariance) Stationarity
A time series is weakly stationary if:
- Mean is constant over time  
- Variance is constant over time  
- Autocovariance depends only on lag  

> Most practical time series analysis focuses on weak stationarity.


## Why Stationarity is Important
- Enables meaningful statistical inference
- Improves model stability
- Required for ARIMA-based models
- Prevents misleading trends and spurious correlations


## How to Test Stationarity

### 1. Visual Inspection
- Plot the time series
- Look for trends, changing variance, or seasonality

**Non-stationary indicators:**
- Upward or downward trend
- Increasing or decreasing variance
- Repeating seasonal patterns


### 2. Rolling Statistics
- Compute rolling mean and rolling variance
- If they change over time, the series is non-stationary


### 3. Augmented Dickey-Fuller (ADF) Test

#### Hypotheses:
- **H₀ (Null):** Series is non-stationary
- **H₁ (Alternative):** Series is stationary

#### Decision Rule:
- If *p-value < 0.05*, reject H₀ → series is stationary


### 4. KPSS Test

#### Hypotheses:
- **H₀ (Null):** Series is stationary
- **H₁ (Alternative):** Series is non-stationary

> ADF and KPSS are often used together for confirmation.


## How to Transform a Non-Stationary Series

### 1. Differencing
Subtract the previous observation from the current one.

- First-order differencing removes trend
- Seasonal differencing removes seasonality

**Example:**  
\[
Y_t' = Y_t - Y_{t-1}
\]


### 2. Log or Power Transformation
Used to stabilize variance.

- Log transform
- Square root transform
- Box-Cox transform

**Example:**  
\[
Y_t' = \log(Y_t)
\]


### 3. Detrending
Remove trend using:
- Regression
- Moving averages


### 4. Seasonal Adjustment
Remove seasonal effects by:
- Seasonal differencing
- Seasonal decomposition


## Summary Table

| Method | Purpose |
|------|--------|
| Differencing | Remove trend and seasonality |
| Log transform | Stabilize variance |
| Detrending | Remove long-term trend |
| Seasonal adjustment | Remove seasonal patterns |

## Question 5
Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models.

## Answer
# Difference Between AR, MA, ARIMA, SARIMA, and SARIMAX Models

Time series models are used to capture patterns such as trend, autocorrelation, and seasonality in time-dependent data. AR, MA, ARIMA, SARIMA, and SARIMAX are progressively more advanced models designed to handle increasing complexity in time series data.

## 1. AR (AutoRegressive) Model

### Definition
The **AR model** predicts the current value of a time series using its **past values**.

### Mathematical Form
\[
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \varepsilon_t
\]

### Key Characteristics
- Uses lagged values of the series
- Assumes stationarity
- Suitable when data shows strong autocorrelation

### Example
Predicting today’s electricity demand based on demand from the previous few hours.


## 2. MA (Moving Average) Model

### Definition
The **MA model** predicts the current value using **past error terms (shocks)** rather than past values.

### Mathematical Form
\[
Y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2} + \dots + \theta_q \varepsilon_{t-q}
\]

### Key Characteristics
- Models random shocks
- Assumes stationarity
- Effective when errors show correlation

### Example
Modeling sudden fluctuations in stock prices caused by unexpected news.

## 3. ARIMA (AutoRegressive Integrated Moving Average)

### Definition
**ARIMA** combines AR and MA models and includes **differencing** to handle non-stationary data.

### Components
- **AR (p):** Autoregressive terms  
- **I (d):** Differencing to achieve stationarity  
- **MA (q):** Moving average terms  

### Mathematical Representation
ARIMA(p, d, q)

### Key Characteristics
- Handles trend
- Requires stationary series after differencing
- No seasonal component

### Example
Forecasting monthly sales data with a long-term upward trend.


## 4. SARIMA (Seasonal ARIMA)

### Definition
**SARIMA** extends ARIMA by explicitly modeling **seasonal patterns** in the data.

### Components
- Non-seasonal: (p, d, q)  
- Seasonal: (P, D, Q, s)

### Mathematical Representation
SARIMA(p, d, q)(P, D, Q, s)

### Key Characteristics
- Handles trend and seasonality
- Seasonal period `s` is fixed (e.g., 12 for monthly data)
- Requires stationary series

### Example
Forecasting electricity demand with daily or yearly seasonal patterns.


## 5. SARIMAX (Seasonal ARIMA with Exogenous Variables)

### Definition
**SARIMAX** further extends SARIMA by including **external (exogenous) variables** that influence the time series.

### Exogenous Variables
- Weather conditions
- Holidays
- Economic indicators

### Key Characteristics
- Models trend, seasonality, and external effects
- Most flexible among ARIMA-based models
- Suitable for real-world forecasting problems

### Example
Forecasting energy demand using temperature and humidity as additional inputs.


## Comparison Table

| Model | Handles Trend | Handles Seasonality | Uses External Variables | Stationarity Required |
|------|--------------|--------------------|-------------------------|----------------------|
| AR | No | No | No | Yes |
| MA | No | No | No | Yes |
| ARIMA | Yes | No | No | After differencing |
| SARIMA | Yes | Yes | No | After differencing |
| SARIMAX | Yes | Yes | Yes | After differencing |

## Question 6
Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset
df = pd.read_csv("AirPassengers.csv")
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# Plot original time series
plt.figure()
plt.plot(df['#Passengers'])
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()

# Decompose the series
decomposition = seasonal_decompose(df['#Passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

## Question 8
Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset
df = pd.read_csv("AirPassengers.csv")
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

ts = df['#Passengers']

# Train SARIMA model
# (p,d,q) = (1,1,1)
# (P,D,Q,s) = (1,1,1,12) for monthly seasonality
model = SARIMAX(
    ts,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Plot results
plt.figure()
plt.plot(ts, label="Observed")
plt.plot(forecast_mean, label="Forecast", linestyle="--")
plt.fill_between(
    conf_int.index,
    conf_int.iloc[:, 0],
    conf_int.iloc[:, 1],
    alpha=0.3
)
plt.title("SARIMA Forecast – AirPassengers (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```

## Question 9
Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate synthetic numerical data
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=1, size=(300, 2))
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))

X = np.vstack((normal_data, outliers))

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
y_pred = lof.fit_predict(X)

# Separate inliers and outliers
inliers = X[y_pred == 1]
detected_outliers = X[y_pred == -1]

# Visualization
plt.figure()
plt.scatter(inliers[:, 0], inliers[:, 1], label="Inliers")
plt.scatter(detected_outliers[:, 0], detected_outliers[:, 1], label="Outliers")
plt.title("Anomaly Detection using Local Outlier Factor (LOF)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
